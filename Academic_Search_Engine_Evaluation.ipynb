# Data Preprocessing

```python
# https://www.kaggle.com/code/matthewmaddock/nlp-arxiv-dataset-transformers-and-umap

# This takes about 1 minute.
import json
import pandas as pd

def setup_dataset(file_name):
    cols = ['id', 'authors', 'title', 'abstract', 'categories', 'comments', 'year']
    data = []

    with open(file_name, encoding='latin-1') as f:
        for line in f:
            doc = json.loads(line)
            year = int('20' + doc['id'][:2]) if doc['id'].startswith('20') or doc['id'].startswith('19') else None
            lst = [doc['id'], doc['authors'], doc['title'], doc['abstract'], doc['categories'], doc['comments'], year]
            data.append(lst)

    df_data = pd.DataFrame(data=data, columns=cols)

    # Filter for documents published on or before YEAR_FILTER
    df_data = df_data[df_data['year'] <= YEAR_FILTER]

    print(df_data.shape)

    return df_data
```

```python
file_name = 'arxiv-metadata-oai-snapshot.json'
df_data = setup_dataset(file_name)
```

```python
def filter_topics(df_data):
  filtered_data = df_data[df_data['categories'].isin(TOPICS)]
  print("filtered_data.shape : ", filtered_data.shape)

  return filtered_data
```

```python
df_data_filtered = filter_topics(df_data)
```

# Data Processing

```python
def get_cat_text(x):

    cat_text = ''

    # Put the codes into a list
    cat_list = x.split(' ')

    for i, item in enumerate(cat_list):

        cat_name = category_map[item]

        # If there was no description available
        # for the category code then don't include it in the text.
        if cat_name != 'Not available':

            if i == 0:
                cat_text = cat_name
            else:
                cat_text = cat_text + ', ' + cat_name

    # Remove leading and trailing spaces
    cat_text = cat_text.strip()

    return cat_text

def clean_text(x):

    # Replace newline characters with a space
    new_text = x.replace("\n", " ")
    # Remove leading and trailing spaces
    new_text = new_text.strip()

    return new_text
```

```python
def preprocess_dataset(df_data):
    df_data['cat_text'] = df_data['categories'].apply(get_cat_text)

    df_data['title'] = df_data['title'].apply(clean_text)
    df_data['abstract'] = df_data['abstract'].apply(clean_text)
    df_data['authors'] = df_data['authors'].apply(clean_text)
    # df_data['comments'] = df_data['comments'].apply(clean_text)


    df_data['prepared_text'] = df_data['title'] + ' \n ' + df_data['abstract']

    return df_data
```

```python
from llama_index.core import Document

def get_arxiv_documents(df_data_filtered):
    df_data = preprocess_dataset(df_data_filtered)

    # arxiv_documents = [Document(text=item) for item in list(df_data['prepared_text'])]
    # arxiv_documents = [Document(text=item) for item in list(df_data['prepared_text'])]
    # metadata={"filename": "<doc_file_name>", "category": "<category>"}

    arxiv_documents = []
    for _, row in df_data.iterrows():
        metadata = {
            "id": row['id'],
            "authors": row['authors'],
            "title": row['title'],
            "category": row['categories'],
            "comments": row['comments'],
            "cat_text": row['cat_text'],
            "prepared_text" : row['prepared_text'],
            "year" : row['year']
        }
        document = Document(text=row['prepared_text'], metadata=metadata)
        arxiv_documents.append(document)

    df_data_final = df_data

    return arxiv_documents, df_data_final
```

```python
def setup_pinecone(index_name, vector_dim, api_key=API_KEY, initial_setup=True):
    pc = Pinecone(api_key=api_key)
    pc.list_indexes()

    # Create an index if it doesn't exist
    if index_name not in pc.list_indexes().names():
        pc.create_index(
          name='my-index',
          dimension=vector_dim,
          metric='cosine',
          spec=PodSpec(
            environment='us-east1-gcp',
            pod_type='p1.x1',
            pods=1
          )
        )

        # wait for index to be initialized
        while not pc.describe_index(index_name).status['ready']:
            time.sleep(1)

    pinecone_index = pc.Index(name=index_name)

    index_stats = pinecone_index.describe_index_stats()

    print("index_stats : \n", index_stats)

    # if initial_setup = True
    # Delete all previous indexes to start fresh
    if initial_setup and (index_stats["total_vector_count"] > 0):
        pinecone_index.delete(delete_all=True)

        index_stats = pinecone_index.describe_index_stats()
        print("index_stats after deleting : \n", index_stats)

    return pinecone_index
```

```python
import os
import urllib
from dateutil import parser
import time
import pandas as pd
```

```python
def fetch_ground_truth(
    ground_truth_file_path
):
    df = pd.read_excel(ground_truth_file_path, engine="openpyxl")

    df['Question'] = df['Questions'].apply(lambda x: x.split('*')[1].strip() if '*' in x else x.strip())

    # Drop the original 'Questions' column and keep only 'Title' and 'First Question'
    df = df[['Title', 'Question']]
    print("Ground Truth DF : ", df)

    return df
```

```python
def get_predictions(query, rerank_postprocessor, top_k):
    results = retrieve_documents(index, query, rerank_postprocessor, top_k=top_k)

    # Compare titles
    return [result["title"] for result in results]
```

```python
# Specify the number of top results to fetch for each query
top_k = 3

ground_truth_file_path = "ground_truth_queries.xlsx"
# Fetch ground truth for each query and store the results
ground_truth_df = fetch_ground_truth(ground_truth_file_path)
# df of "query" - > "title"

# Fetch predictions for each query and store the results
predictions_data = {}
for query in ground_truth_df["Question"].to_list():
    predictions = get_predictions(query, rerank_postprocessor, top_k)
    # Store "query" : top_k titles in the predictions data
    predictions_data[query] = predictions
```

```python
from collections import defaultdict

def evaluate_model(ground_truth_df, predictions_data):
  # Convert ground truth DataFrame to a dictionary for easier access
  ground_truth_dict = dict(zip(ground_truth_df['Question'], ground_truth_df['Title']))

  # Variables to keep track of matches and relevant documents
  true_positives = 0
  total_relevant = len(ground_truth_dict)  # Since we're checking the top-1, this is the total queries
  total_retrieved = len(predictions_data)  # Assuming each query has a top-1 prediction

  # Iterate over predictions to calculate true positives
  for query, predictions in predictions_data.items():
      if query in ground_truth_dict:
          # Check if top-1 prediction matches ground truth
          # if predictions[0] == ground_truth_dict[query]:
          if ground_truth_dict[query] in predictions:
              true_positives += 1

  # Calculate precision and recall
  precision = true_positives / total_retrieved if total_retrieved else 0
  recall = true_positives / total_relevant if total_relevant else 0

  print("Precision : ", precision)
  print("Recall : ", recall)
```

```python
evaluate_model(ground_truth_df, predictions_data)
```
